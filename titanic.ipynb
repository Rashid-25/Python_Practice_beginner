```python

import numpy as np
import matplotlib.pyplot as plt

# Define a vector
v = np.array([2, 3])

# Define a transformation matrix (Scaling by 2x and 3x)
scaling_matrix = np.array([[2, 0], [0, 3]])
v_scaled = scaling_matrix @ v

# Define a rotation matrix (Rotating by 45 degrees)
theta = np.radians(45)  # Convert degrees to radians
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])
v_rotated = rotation_matrix @ v

# Plot the vectors before and after transformation
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label="Original Vector")
plt.quiver(0, 0, v_scaled[0], v_scaled[1], angles='xy', scale_units='xy', scale=1, color='g', label="Scaled Vector")
plt.quiver(0, 0, v_rotated[0], v_rotated[1], angles='xy', scale_units='xy', scale=1, color='b', label="Rotated Vector")

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid()
plt.legend()
plt.title("Vector Transformations")
plt.show()
```
