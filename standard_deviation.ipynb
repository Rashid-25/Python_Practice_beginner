# standard deviation

```python

import numpy as np
import matplotlib.pyplot as plt

data = np.array([2, 3, 4, 5, 6])
mean = data.mean()
std_dev = data.std(ddof=1)

plt.figure(figsize=(10, 4))
plt.scatter(data, [1]*len(data), color='blue', label='Data Points')
plt.axvline(mean, color='red', linestyle='dashed', label=f'Mean = {mean}')
plt.axvline(mean - std_dev, color='green', linestyle='dotted', label=f'-1 SD = {mean - std_dev:.2f}')
plt.axvline(mean + std_dev, color='green', linestyle='dotted', label=f'+1 SD = {mean + std_dev:.2f}')
plt.legend()
plt.title('Mean and Standard Deviation Visualization')
plt.xlabel('Data Values')
plt.grid(True)
plt.show()
```

# standard error

```python


data = np.array([2, 3, 4, 5, 6])
std_dev = data.std(ddof=1)  # Sample Standard Deviation
n = len(data)
std_error = std_dev / np.sqrt(n)

print(f"Standard Error = {std_error:.2f}")
```
