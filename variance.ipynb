# First dataset

```python
import numpy as np

df = np.array([2, 3, 4, 5, 6])  # Define the array

mean = np.mean(df)  # Calculate the mean
variance = np.var(df, ddof=1)

print("Mean:", mean)
print("Variance:", variance)
```

# Second dataset

```python

import numpy as np
import matplotlib.pyplot as plt

# Dataset
marks = np.array([45, 50, 55, 60, 65])

# Calculate Mean
mean = np.mean(marks)

# Calculate Sample Variance
sample_variance = np.var(marks, ddof=1)  # ddof=1 for Sample Variance
print("Mean:", mean)
print("Sample Variance:", sample_variance)

# Visualization
plt.figure(figsize=(8, 5))
plt.bar(range(1, len(marks) + 1), marks, color='skyblue', label='Marks')
plt.axhline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.1f}')

# Adding Labels
plt.title("Marks with Mean Line")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.legend()
plt.grid(True)

# Show Plot
plt.show()
```

# Using same dataset but creating histogram

```python

import numpy as np
import matplotlib.pyplot as plt

# Dataset
marks = np.array([45, 50, 55, 60, 65])

# Calculate Mean and Sample Variance
mean = np.mean(marks)
sample_variance = np.var(marks, ddof=1)

print("Mean:", mean)
print("Sample Variance:", sample_variance)

# Plot Histogram
plt.figure(figsize=(8, 5))
plt.hist(marks, bins=5, color='skyblue', edgecolor='black', alpha=0.7, label='Marks')

# Plot Mean Line
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.1f}')

# Add Labels
plt.title("Histogram of Marks with Mean Line")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)

# Show Plot
plt.show()
```
